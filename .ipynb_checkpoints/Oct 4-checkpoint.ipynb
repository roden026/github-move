```python
# split apply combine - using red wine data from above
dframe_wine.head()
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort_values('alcohol', ascending = False, inplace=True) #inplace makes it effect the dataframe 
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker) # applies function 1 by 1 to each group
dframe_wine.head()
```

```python
num_of_qual = dframe_wine['quality'].value_counts()
```

```python
num_of_qual
```

```python
dframe_wine[dframe_wine.alc_content_rank==1].head(len(num_of_qual))
```

```python
# Cross Tabs - basically a special pivot table
from StringIO import StringIO
```

```python
data="""\
Sample Animal Intelligence
1 dog smart
2 dog smart
3 cat dumb
4 cat dumb
5 dog dumb
6 cat smart"""
```

```python
dframe = pd.read_table(StringIO(data), sep='\s+')
```

```python
dframe
```

```python
pd.crosstab(dframe.Animal, dframe.Intelligence, dframe.Sample, aggfunc=np.sum, margins=True)
```

```python
# read in pokedata
location = "C:\Users\erroden\Documents\Python Scripts\Test Data\pokemon.csv"

pokedata = pd.read_csv(location)
pokedata['total_power'] = pokedata[['hp', 'attack', 'defense', 'speed']].sum(axis=1)
```
